```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("forest_fire_data.csv")
df
```

```python
def extract_row_with_constraint(df, category):
    df.loc[df['Category'] == category] = df.loc[df['Category'] == category].fillna(0)
    return df.loc[df['Category'] == category]
```

```python
x = range(1983, 2020)
yfires = np.array(extract_row_with_constraint(df, "Fires").iloc[:, -(2019-1983+1):]).flatten()
```

```python
for i in range(len(yfires)):
    yfires[i] = yfires[i].replace(',', '')
    yfires[i] = int(yfires[i])
```

```python
plt.plot(x, yfires)
plt.title("Number of forest fires vs Time")
plt.xlabel("Time (year)")
plt.ylabel("Number of forest fires in United States")
plt.savefig("fires.png", bbox_inches='tight')
```

```python
yacres = np.array(extract_row_with_constraint(df, "Acres").iloc[:, -(2019-1983+1):]).flatten()
```

```python
for i in range(len(yacres)):
    yacres[i] = yacres[i].replace(',', '')
    yacres[i] = int(yacres[i])
```

```python
plt.plot(x, yacres)
plt.title("Number of acres affected by forest fires vs Time")
plt.xlabel("Time (year)")
plt.ylabel("Number of acres affected by forest fires in United States")
plt.savefig("acres.png", bbox_inches='tight')
```
